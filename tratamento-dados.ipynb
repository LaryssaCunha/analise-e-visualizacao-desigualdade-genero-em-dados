# **Análise Exploratória:**
- O objetivo principal nessa etapa era verificar qual era a real disparidade entre a quantidade de homens e mulheres na área de dados e como ela ocorria ao longo dos anos de 2021, 2022 e 2023.

 **Primeira etapa:**
- Contar quantidade de mulheres e homens nos anos de 2021, 2022 e 2023. 

```python
#contando só as coluna de gênero feminino em 2022
quantidade_mulheres_2022 = df_std2022[df_std2022['genero'] == 'Feminino'].count()
quantidade_mulheres_2022 #total de mulheres
```

```python
# Contando apenas o número de ocorrências de "Feminino" na coluna "genero"
quantidade_mulheres_2022 = df_std2022[df_std2022['genero'] == 'Feminino'].shape[0]
quantidade_mulheres_2022
```

```python
quantidade_mulheres_2023 = df_std2023[df_std2023['genero'] == 'Feminino'].count()
quantidade_mulheres_2023
```

```python
quantidade_mulheres_2023 = df_std2023[df_std2023['genero'] == 'Feminino'].shape[0]
quantidade_mulheres_2023
```

```python
#Contabilizando a diferença absoluta entre o total de mulheres de 2022 e 2023
diferenca_absoluta = quantidade_mulheres_2023 - quantidade_mulheres_2022
diferenca_absoluta
```

```python
#Valor do total de mulheres de 2022 ( = 1056) em porcentagem em relação ao total
porcentagem_mulheres_2022 = (quantidade_mulheres_2022 / df_std2022["genero"].count()) * 100
porcentagem_mulheres_2022
```

```python
#Valor total de mulheres de 2023 ( = 1293 ) em porcentagem em relação ao total
porcentagem_mulheres_2023 = (quantidade_mulheres_2023 / df_std2023["genero"].count()) * 100
porcentagem_mulheres_2023
```

```python
diferenca_percentual = porcentagem_mulheres_2022 - porcentagem_mulheres_2023
f'{diferenca_percentual:.2f}%'
```

# **Visualização dos dados.**

```python
#Dados
categorias1 = ["Mulheres em Dados em 2022", "Mulheres em Dados em 2023"]
valores = [quantidade_mulheres_2022, quantidade_mulheres_2023]

#Criando o g´rafico de barras
plt.figure(figsize=(8, 6)) #Ajuste o tamanho da figura
barras = plt.bar(categorias1, valores, color='#FF69B4', edgecolor = 'black')

#Ajustando o eixo Y para mostrar apenas entre 100 e 1500
plt.ylim(1000, 1350)

#Adicionando os rótulos acima das barras
for barra in barras:
    altura = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width() / 2, altura + 10, #Posição acima da barra
        f'{altura}', #Valor da barra
        ha='center', va='bottom', fontsize=12, color='black', fontweight='bold'            
    )

#Ajustando os rótulos e os títulos
plt.title("Comparação do total de mulheres em dados (2022 vs 2023)", fontsize=14, fontweight='bold')
plt.ylabel("Quantidade", fontsize=12)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

#Mostrando o gráfico
plt.tight_layout()
plt.show()
```

```python
#Dados
categorias2 = ["Porcentagem de mulheres em dados em 2022", "Porcentagem de mulheres em dados em 2023"]
valores2 = [porcentagem_mulheres_2022, porcentagem_mulheres_2023]

#Criando o gráfico de barras
plt.figure(figsize=(12, 7)) #Ajusta o tamanho da figura
barras2 = plt.bar(categorias2, valores2, color="#FF69B4", edgecolor='black')

#Ajustando o eixo Y para mostrar apenas entre 23% e 25%
plt.ylim(24, 25)

#Adicionado os rótulos acima das barras para destacar os valores
for barra2 in barras2:
    altura2 = barra2.get_height()
    plt.text(
        barra2.get_x() + barra2.get_width() / 2, altura2 + 0.1, #Posição acima da barra
        f'{altura2:.2f}%', #Valor da barra com duas casas decimais
        ha='center', va='bottom', fontsize=16, color='black', fontweight='bold'
    )

#Ajustando os rótulos e o título
plt.title("Comparação da Porcentagem de Mulheres em Dados (2022 vs 2023)", fontsize=16, fontweight='bold')
plt.ylabel("Porcentagem (%)", fontsize = 14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

#Ajustar as margens
plt.subplots_adjust(top=0.9, bottom=0.15)

#Exibindo o gráfico
plt.tight_layout()
plt.show()
```
