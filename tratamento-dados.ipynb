# **Análise Exploratória:**
- O objetivo principal nessa etapa era verificar qual era a real disparidade entre a quantidade de homens e mulheres na área de dados e como ela ocorria ao longo dos anos de 2021, 2022 e 2023.

 **Primeira etapa:**
- Contar quantidade de mulheres e homens nos anos de 2021, 2022 e 2023. 

```python
#contando só as coluna de gênero feminino em 2021, 2022 e 2023
quantidade_mulheres_2022 = std2022_tratado[std2022_tratado['Gênero'] == 'Feminino']["Gênero"].count()
quantidade_mulheres_2022 #total de mulheres
```

```python
quantidade_mulheres_2023 = std2023_tratado[std2023_tratado['Gênero'] == 'Feminino']["Gênero"].count()
quantidade_mulheres_2023
```

```python
#transformando esses valores para porcentagem
porcentagem_mulheres_2022 = (quantidade_mulheres_2022 / std2022_tratado["Gênero"].count()) * 100
porcentagem_mulheres_2022
```

```python
porcentagem_mulheres_2023 = (quantidade_mulheres_2023 / std2023_tratado["Gênero"].count()) * 100
porcentagem_mulheres_2023
```

# **Visualização dos dados.**

```python
plt.bar(["Mulheres em Dados em 2022", "Mulheres em Dados em 2023"], [quantidade_mulheres_2022, quantidade_mulheres_2023])
#inserir aqui dos homens ao lado
```

```python
plt.bar(["Porcentagem de mulheres em dados em 2022", "Porcentagem de mulheres em dados em 2023"], [porcentagem_mulheres_2022, porcentagem_mulheres_2023])
#vou perguntar a edu do rotacionar o grafico pra acertar
```
